```python
import os

# Path to your labels folder (do this for both 'train' and 'val' folders)
label_dir = "datasets/leaf_data/labels/train"

def convert_labels_to_zero(directory):
    # List all .txt files in the directory
    txt_files = [f for f in os.listdir(directory) if f.endswith('.txt')]
    
    if not txt_files:
        print(f"No text files found in {directory}")
        return

    for filename in txt_files:
        file_path = os.path.join(directory, filename)
        
        with open(file_path, 'r') as f:
            lines = f.readlines()
        
        new_lines = []
        for line in lines:
            parts = line.split()
            if len(parts) > 0:
                # Force the first element (class index) to '0'
                # parts[1:] keeps the coordinates (x, y, w, h) exactly as they are
                parts[0] = '0'
                new_lines.append(" ".join(parts) + "\n")
        
        # Overwrite the file with the corrected labels
        with open(file_path, 'w') as f:
            f.writelines(new_lines)

    print(f"Successfully processed {len(txt_files)} files in {directory}")

# Run the function
convert_labels_to_zero(label_dir)
# If you have a val folder, run it there too:
# convert_labels_to_zero("datasets/leaf_data/labels/val")
```
