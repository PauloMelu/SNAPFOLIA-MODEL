```python
from ultralytics import YOLO

# 1. Load the YOLO26 model
# You can use yolo26n.pt (nano), yolo26s.pt (small), yolo26m.pt (medium), etc.
model = YOLO("yolo26n.pt")

# 2. Train the model
# 'data' points to your dataset.yaml file (see step 3)
results = model.train(
    data="custom_data.yaml", 
    epochs=100,              # Number of training epochs
    imgsz=640,               # Input image size
    device=0,                # GPU device index (use 'cpu' if no GPU)
    optimizer="MuSGD",       # New hybrid optimizer introduced in YOLO26
    project="my_yolo26_run", # Project folder name
    name="experiment_1"      # Specific run name
)

# 3. Validate the model
metrics = model.val()

# 4. Export the model for deployment (NMS-free)
model.export(format="onnx")
```
